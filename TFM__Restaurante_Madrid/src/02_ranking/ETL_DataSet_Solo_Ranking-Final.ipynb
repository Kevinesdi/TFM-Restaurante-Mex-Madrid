RANKING

```python
#NORMALIZAMOS TODO (0â€“1)

#competencia se invierte â†’ menos restaurantes = mejor
#esto ya estÃ¡ alineado con mexicano

rank_df = df.copy()

def norm(s):
    s = pd.to_numeric(s, errors="coerce")
    if s.dropna().empty or s.max() == s.min():
        return s*0
    return (s - s.min()) / (s.max() - s.min())


rank_df["footfall_norm"] = norm(rank_df["footfall_day_avg"])
rank_df["access_norm"]   = norm(rank_df["intermodal_score"])

# ðŸ”‘ COMPETENCIA (usar rest_density_500m, no res_density_500m)
rank_df["competition_norm"] = norm(
    1 / (1 + rank_df["rest_density_500m"])
)

rank_df["mex_niche_norm"] = norm(rank_df["footfall_x_mex_share"])
rank_df["resident_norm"]  = norm(rank_df["res_density"])

#DEFINIMOS EL Mexican Location Score

#Este score: va de 0 a 1
#NO es un KPI financiero
#es un ranking estratÃ©gico

rank_df["mexican_location_score"] = (
    0.30 * rank_df["footfall_norm"] +
    0.25 * rank_df["mex_niche_norm"] +
    0.20 * rank_df["access_norm"] +
    0.15 * rank_df["competition_norm"] +
    0.10 * rank_df["resident_norm"]
)


#CREAMOS EL RANKING FINAL

#Top 15 mejores ubicaciones para abrir un restaurante mexicano en Madrid

ranking = (
    rank_df[[
        "grid_id",
        "lat_center",
        "lon_center",
        "mexican_location_score",
        "footfall_day_avg",
        "footfall_x_mex_share",
        "rest_density_500m",
        "intermodal_score"
    ]]
    .sort_values("mexican_location_score", ascending=False)
    .reset_index(drop=True)
)

ranking.head(15)
```
