```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
usuario = 'postgres'
contraseña = 'postgre'
host = 'localhost'
puerto = '5432'
base_de_datos = 'postgres'
```

```python
conexion = f'postgresql+psycopg2://{usuario}:{contraseña}@{host}:{puerto}/{base_de_datos}'
```

```python
conn = create_engine(conexion)
```

```python
conn
```

```python
try:
    version = pd.read_sql("SELECT version();", conn).iloc[0, 0]
    print("Conexión exitosa con SQLAlchemy.")
    print(f"Versión del servidor PostgreSQL: {version}")
except Exception as e:
    print(f"Ocurrió un error al conectar con SQLAlchemy: {e}")
```

```python
consulta_1 = """
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'dw_tfm_restaurantes'
ORDER BY table_name;
"""
pd.read_sql(consulta_1, conn)
```

```python
consulta_2 = """
SELECT *
FROM 
  information_schema.columns 
WHERE 
  table_name = 'terrazas';
"""
pd.read_sql(consulta_2, conn)
```

```python
consulta_3= """
SELECT 
  column_name, 
  data_type, 
  character_maximum_length, 
  is_nullable, 
  column_default 
FROM 
  information_schema.columns 
WHERE 
  table_name = 'terrazas'
ORDER BY
    ordinal_position
"""

pd.read_sql(consulta_3, conn)
```

Volcado Datos

```python
for col in df_terrazas.columns:
        if df_terrazas[col].dtype == 'object':
            # 1. Corregir el error "0.." reemplazándolo por "0."
            df_terrazas[col] = df_terrazas[col].str.replace('0..', '0.', regex=False)
```

```python
# --- LIMPIEZA ESPECÍFICA DE TIPOS ---
# Intentamos convertir a numérico las columnas que sabemos que son coordenadas
coord_cols = ['Latitud', 'longitud', 'Longitud', 'coordenada_x_local', 'coordenada_y_local']
for col in coord_cols:
        if col in df_terrazas.columns:
            df_terrazas[col] = pd.to_numeric(df_terrazas[col], errors='coerce') # Si algo falla, pone NaN (NULL en SQL)
```

```python
df_terrazas.to_sql('terrazas', schema='dw_tfm_restaurantes', con = conn, if_exists = 'append', index=False)
```

```python
count_db = pd.read_sql("SELECT COUNT(*) AS count FROM dw_tfm_restaurantes.terrazas", conn).iloc[0, 0]
count_df = len(df_terrazas)

print(f"Filas en DataFrame: {count_df}")
print(f"Filas en tabla dw_aero.dm_pais: {count_db}")

if count_df == count_db:
    print("La carga se efectuó correctamente.")
else:
    print("Hay discrepancias en la cantidad de filas insertadas.")
```
