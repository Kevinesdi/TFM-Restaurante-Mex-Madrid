```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
usuario = 'postgres'
contraseña = 'postgre'
host = 'localhost'
puerto = '5432'
base_de_datos = 'postgres'
```

```python
conexion = f'postgresql+psycopg2://{usuario}:{contraseña}@{host}:{puerto}/{base_de_datos}'
```

```python
conn = create_engine(conexion)
```

```python
df_licencias_procesado = df_licencias.copy()
```

```python
import numpy as np 

# Valores a reemplazar
valores_a_reemplazar = [
    99999999999999,
    9999999999,
    'SIN INFORMACION',
    9999,
    999999.999999,
    999999.99, # Si hubiera alguna variación
    99
] 
df_licencias_procesado.replace(valores_a_reemplazar, np.nan, inplace=True)

# Después de reemplazar, convertir las columnas a su tipo de dato correcto

df_licencias_procesado['id_local'] = pd.to_numeric(df_licencias_procesado['id_local'], errors='coerce')
df_licencias_procesado['id_distrito_local'] = pd.to_numeric(df_licencias_procesado['id_distrito_local'], errors='coerce')
df_licencias_procesado['id_barrio_local'] = pd.to_numeric(df_licencias_procesado['id_barrio_local'], errors='coerce')
df_licencias_procesado['coordenada_x_local'] = pd.to_numeric(df_licencias_procesado['coordenada_x_local'], errors='coerce')
df_licencias_procesado['coordenada_y_local'] = pd.to_numeric(df_licencias_procesado['coordenada_y_local'], errors='coerce')
```

```python
df_licencias_procesado[df_licencias_procesado.duplicated(subset='id_local', keep= False)].sort_values(by = 'id_local')
```

```python
df_licencias_procesado.dropna(subset=['id_local'], inplace=True)
```

```python
df_licencias_procesado.duplicated().values.any()
```

```python
df_licencias_procesado[df_licencias_procesado.duplicated(subset='id_local', keep= False)].sort_values(by = 'id_local')
```

```python
#Reemplazar valores con errores en latitud y longitud
for col in df_licencias_procesado.columns:
        if df_licencias_procesado[col].dtype == 'object':
            # 1. Corregir el error "0.." reemplazándolo por "0."
            df_licencias_procesado[col] = df_licencias_procesado[col].str.replace('0..', '0.', regex=False)
```

```python
# --- LIMPIEZA ESPECÍFICA DE TIPOS ---
# Intentamos convertir a numérico las columnas que sabemos que son coordenadas
coord_cols = ['Latitud', 'longitud', 'Longitud', 'coordenada_x_local', 'coordenada_y_local']
for col in coord_cols:
        if col in df_licencias_procesado.columns:
            df_licencias_procesado[col] = pd.to_numeric(df_licencias_procesado[col], errors='coerce') # Si algo falla, pone NaN (NULL en SQL)
```

Volcado Datos

```python
df_licencias_procesado.to_sql('licencias', schema='dw_tfm_restaurantes', con = conn, if_exists = 'append', index=False)
```

```python
count_db = pd.read_sql("SELECT COUNT(*) AS count FROM dw_tfm_restaurantes.licencias", conn).iloc[0, 0]
count_df = len(df_licencias_procesado)

print(f"Filas en DataFrame: {count_df}")
print(f"Filas en tabla dw_tfm_restaurantes: {count_db}")

if count_df == count_db:
    print("La carga se efectuó correctamente.")
else:
    print("Hay discrepancias en la cantidad de filas insertadas.")
```
