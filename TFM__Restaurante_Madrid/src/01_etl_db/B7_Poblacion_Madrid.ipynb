```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
usuario = 'postgres'
contraseña = 'postgre'
host = 'localhost'
puerto = '5432'
base_de_datos = 'postgres'
```

```python
conexion = f'postgresql+psycopg2://{usuario}:{contraseña}@{host}:{puerto}/{base_de_datos}'
```

```python
conn = create_engine(conexion)
```

```python
conn
```

```python
try:
    version = pd.read_sql("SELECT version();", conn).iloc[0, 0]
    print("Conexión exitosa con SQLAlchemy.")
    print(f"Versión del servidor PostgreSQL: {version}")
except Exception as e:
    print(f"Ocurrió un error al conectar con SQLAlchemy: {e}")
```

```python
consulta_1 = """
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'dw_tfm_restaurantes'
ORDER BY table_name;
"""
pd.read_sql(consulta_1, conn)
```

```python
consulta_2 = """
SELECT *
FROM 
  information_schema.columns 
WHERE 
  table_name = 'poblacion_madrid';
"""
pd.read_sql(consulta_2, conn)
```

```python
consulta_3= """
SELECT 
  column_name, 
  data_type, 
  character_maximum_length, 
  is_nullable, 
  column_default 
FROM 
  information_schema.columns 
WHERE 
  table_name = 'poblacion_madrid'
ORDER BY
    ordinal_position
"""

pd.read_sql(consulta_3, conn)
```

Eliminar los nulos específicos en las columnas de conteo

```python
columnas_poblacion = ['num_personas', 'num_personas_hombres', 'num_personas_mujeres']
df_poblacion_madrid_sin_totales = df_poblacion_madrid_sin_totales.dropna(subset=columnas_poblacion)
```

Volcado Datos

```python
df_poblacion_madrid.to_sql('poblacion_madrid', schema='dw_tfm_restaurantes', con = conn, if_exists = 'append', index=False)
```

```python
count_db = pd.read_sql("SELECT COUNT(*) AS count FROM dw_tfm_restaurantes.poblacion_madrid", conn).iloc[0, 0]
count_df = len(df_poblacion_madrid)

print(f"Filas en DataFrame: {count_df}")
print(f"Filas en tabla dw_tfm_restaurantes: {count_db}")

if count_df == count_db:
    print("La carga se efectuó correctamente.")
else:
    print("Hay discrepancias en la cantidad de filas insertadas.")
```
