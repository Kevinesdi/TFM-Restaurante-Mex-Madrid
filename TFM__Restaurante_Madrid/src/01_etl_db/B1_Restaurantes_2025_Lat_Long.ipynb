```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
usuario = 'postgres'
contraseña = 'postgre'
host = 'localhost'
puerto = '5432'
base_de_datos = 'postgres'
```

```python
conexion = f'postgresql+psycopg2://{usuario}:{contraseña}@{host}:{puerto}/{base_de_datos}'
```

```python
conn = create_engine(conexion)
```

```python
conn
```

```python
try:
    version = pd.read_sql("SELECT version();", conn).iloc[0, 0]
    print("Conexión exitosa con SQLAlchemy.")
    print(f"Versión del servidor PostgreSQL: {version}")
except Exception as e:
    print(f"Ocurrió un error al conectar con SQLAlchemy: {e}")
```

```python
consulta_1 = """
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'dw_tfm_restaurantes'
ORDER BY table_name;
"""
pd.read_sql(consulta_1, conn)
```

```python
consulta_2 = """
SELECT *
FROM 
  information_schema.columns 
WHERE 
  table_name = 'restaurantes';
"""
pd.read_sql(consulta_2, conn)
```

```python
consulta_3= """
SELECT 
  column_name, 
  data_type, 
  character_maximum_length, 
  is_nullable, 
  column_default 
FROM 
  information_schema.columns 
WHERE 
  table_name = 'restaurantes'
ORDER BY
    ordinal_position
"""

pd.read_sql(consulta_3, conn)
```

Duplicados para eliminar

```python
df_restaurantes[df_restaurantes.duplicated(subset='id_local', keep= False)].sort_values(by = 'id_local')
```

```python
df_restaurantes = df_restaurantes.drop_duplicates(subset='id_local', keep='first')
```

```python
df_restaurantes.duplicated().values.any()
```

```python
df_restaurantes_procesado = df_restaurantes.copy()
```

```python
import numpy as np # Necesario para np.nan

# Suponiendo que tu DataFrame se llama df_tu_dataset
# Reemplaza 'df_tu_dataset' con el nombre real de tu DataFrame (ej. df_actividad_economica2025)

# Define los valores de centinela que quieres reemplazar
valores_a_reemplazar = [
    99999999999999,
    9999999999,
    'SIN INFORMACION',
    9999,
    999999.999999,
    999999.99, # Si hubiera alguna variación
    99
] 
df_restaurantes_procesado.replace(valores_a_reemplazar, np.nan, inplace=True)



# Después de reemplazar, es una buena idea intentar convertir las columnas a su tipo de dato correcto
# Por ejemplo, las coordenadas y IDs a numérico
df_restaurantes_procesado['id_local'] = pd.to_numeric(df_restaurantes_procesado['id_local'], errors='coerce')
df_restaurantes_procesado['id_distrito_local'] = pd.to_numeric(df_restaurantes_procesado['id_distrito_local'], errors='coerce')
df_restaurantes_procesado['id_barrio_local'] = pd.to_numeric(df_restaurantes_procesado['id_barrio_local'], errors='coerce')
df_restaurantes_procesado['coordenada_x_local'] = pd.to_numeric(df_restaurantes_procesado['coordenada_x_local'], errors='coerce')
df_restaurantes_procesado['coordenada_y_local'] = pd.to_numeric(df_restaurantes_procesado['coordenada_y_local'], errors='coerce')

```

```python
df_restaurantes_procesado.dropna(subset=['id_local'], inplace=True)
```

```python
df_restaurantes_procesado.dropna(subset=['id_seccion'], inplace=True)
```

```python
df_restaurantes_procesado[df_restaurantes_procesado.duplicated(subset='id_seccion', keep= False)].sort_values(by = 'id_seccion')
```

Volcado Datos

```python
df_restaurantes_procesado.to_sql('restaurantes', schema='dw_tfm_restaurantes', con = conn, if_exists = 'append', index=False)
```

```python
count_db = pd.read_sql("SELECT COUNT(*) AS count FROM dw_tfm_restaurantes.restaurantes", conn).iloc[0, 0]
count_df = len(df_restaurantes_procesado)

print(f"Filas en DataFrame: {count_df}")
print(f"Filas en tabla dw_tfm_restaurantes: {count_db}")

if count_df == count_db:
    print("La carga se efectuó correctamente.")
else:
    print("Hay discrepancias en la cantidad de filas insertadas.")
```
